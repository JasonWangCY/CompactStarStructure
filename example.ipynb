```python
df2 = pd.read_csv('app/output/MRrelation_WD.csv')
```

```python
totR_list = df2['totR_list'].values
totM_list = df2['totM_list'].values
totR_list_TOV = df2['totR_list_TOV'].values
totM_list_TOV = df2['totM_list_TOV'].values
```

```python
plt.rcParams['figure.figsize'] = [10, 10]
fig, ax = plt.subplots()
ax.plot(totM_list, totR_list, c='b', label='Newtonian')
ax.plot(totM_list_TOV, totR_list_TOV, c='r', label='TOV')
ax.set_title('Mass-radius relationship of WD')
ax.set_ylabel(r'$r$ ($R_\odot$)')
ax.set_xlabel(r'$M$ ($M_\odot$)')
ax.tick_params(axis='both', which='major', direction = 'in')
ax.legend()
plt.show()
```
