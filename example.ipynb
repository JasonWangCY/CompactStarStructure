```python
import pandas as pd
import matplotlib.pyplot as plt
import StellarStructure.functions as fn
import numpy as np
```

```python
df = pd.read_csv('app/output/MRrelation_WD_dm380_rho5e5.csv', sep=',')
```

```python
df_exact = pd.read_csv('app/output/MRrelation_exact_WD_dm380_rho1e12.csv', sep=',')
```

```python
totR_dm_exact = df_exact['totR_list_TOV'].values
totM_dm_exact = df_exact['totM_list_TOV'].values
```

```python
totR_dm_exact = totR_dm_exact[totR_dm_exact>1e-5]
totM_dm_exact = totM_dm_exact[totM_dm_exact>1e-5]
```

```python
totR_dm = df['totR_list_TOV'].values
totM_dm = df['totM_list_TOV'].values
max(totM_dm)
```

```python
df2 = pd.read_csv('app/output/MRrelation_WD.csv', sep=',')
```

```python
totR = df2['totR_list_TOV'].values
totM = df2['totM_list_TOV'].values
```

```python
plt.rcParams['figure.figsize'] = [10, 10]
fig, ax = plt.subplots()
ax.plot(totM, totR, c='b', label='w/o dm')
ax.plot(totM_dm_exact, totR_dm_exact, c='r', label='w/ dm')
ax.set_title('Mass-radius relationship of WD')
ax.set_ylabel(r'$r$ ($R_\odot$)')
ax.set_xlabel(r'$M$ ($M_\odot$)')
ax.tick_params(axis='both', which='major', direction = 'in')
ax.legend()
plt.show()
```
